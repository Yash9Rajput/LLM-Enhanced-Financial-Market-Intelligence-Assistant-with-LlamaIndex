```python
import os
import urllib.request
import zipfile
```

```python
data_url = "https://raw.githubusercontent.com/entbappy/Branching-tutorial/master/articles.zip"
```

```python
!mkdir my_repo_files
```

```python
output_dir = "/content/my_repo_files"
```

```python
def download_and_extract():

  #if dir doesn't exist
  os.makedirs(output_dir, exist_ok=True)

  #download zip file
  filename , headers = urllib.request.urlretrieve(
      url = data_url,
      filename = "articles.zip"
  )

  print(f"Downloaded {filename} with headers {headers}")

  #extract zip file
  with zipfile.ZipFile(filename, 'r') as zip_ref:
    zip_ref.extractall(output_dir)
  print("files extracted successfully")

  #remove zip file
  os.remove(filename)
  print("zip file removed successfully")


download_and_extract()
```

```python
import streamlit as st
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex

# Load data and build query engine once
@st.cache_resource
def load_engine():
    documents = SimpleDirectoryReader("/content/my_repo_files/articles").load_data()
    index = VectorStoreIndex.from_documents(documents)
    return index.as_query_engine()

query_engine = load_engine()

st.set_page_config(page_title="Financial Stock Analysis Chatbot", layout="centered")
st.title("📊 Financial Stock Analysis Chatbot (LlamaIndex)")
st.markdown("Ask me anything about **stock market trends, predictions, or analysis.**")

query = st.text_input("💬 Enter your question:")

if query:
    with st.spinner("Analyzing stock data..."):
        response = query_engine.query(query)
    st.write("### ✅ Answer:")
    st.success(str(response))
```

```python
%%writefile app.py
import streamlit as st

st.set_page_config(page_title="Financial Stock Analysis Chatbot", layout="centered")
st.title("📊 Financial Stock Analysis Chatbot (LlamaIndex)")
st.markdown("Ask me anything about **stock market trends, predictions, or analysis.**")

query = st.text_input("💬 Enter your question:")

def get_bot_response(user_query):
    return f" [Simulated Answer] You asked: '{user_query}'.\n\nMarket data analysis feature coming soon!"

if query:
    with st.spinner("Analyzing stock data..."):
        response = get_bot_response(query)
    st.write("###  Answer:")
    st.success(response)

st.markdown("---")
st.markdown(" *Powered by LlamaIndex + Streamlit*")
```
